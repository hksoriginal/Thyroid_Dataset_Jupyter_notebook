```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hypothyroid.csv')
```

```python
df.info()
```

```python
df.head(10)
```

##  Replacing Categorical features with Numerical Data & NaN Values

```python
df.replace({'f': 0, 't': 1,'?': np.nan}, inplace=True)
```

```python
df["sex"].replace({"F": 1, "M": 0}, inplace=True)
```

```python
df["binaryClass"].replace({"P": 1, "N": 0}, inplace=True)
```

```python
df.head(10)
```

```python
fig, ax = plt.subplots(dpi = 150)
sns.heatmap(df.isnull(),cbar = False,ax =ax,cmap= 'viridis')
```

```python
df.isnull().sum()
```

## Dropping Columns

```python
df.drop(['TBG','TBG measured','referral source'], axis = 1,inplace = True)
```

## Dropping NaN values

```python
df1 = df.dropna(subset=['sex','age'])
df1.head(-10)
```

```python
fig, ax = plt.subplots(dpi = 120, figsize= (25,15))
sns.heatmap(df1.corr(),annot=True,ax =ax)
```

## Typcasting columns

```python
df1[['age','T3','TT4','T4U','FTI']]=df1[['age','T3','TT4','T4U','FTI']].astype(float)
```

```python
fig, ax = plt.subplots(dpi = 120, figsize= (25,15))
sns.heatmap(df1.corr(),annot=True,ax =ax)
```

```python
df1.drop(['T4U measured'],axis=1,inplace=True)
```

```python
fig, ax = plt.subplots(dpi = 150)
sns.heatmap(df1.isnull(),cbar = False,ax =ax,cmap= 'viridis')
```

## Separating Features and Labels

```python
features = df.drop('binaryClass',axis=1)
label = df['binaryClass']
```

```python
label
```

# Passing through Pipeline

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
my_pipe = Pipeline([('imputer', SimpleImputer(strategy='mean'))])
```

```python
features_tr = my_pipe.fit_transform(features)
features_tr
```

```python
features_tr_df = pd.DataFrame(features_tr)
features_tr_df
```

```python
fig, ax = plt.subplots(dpi = 100)
sns.heatmap(features_tr_df.isnull(),cbar = False,ax =ax,cmap= 'viridis')
```

# Train Test Splitting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(features_tr,label,test_size=0.2,random_state=42)
```

```python
x_test
```

# Model Selection & Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```



### Random Forest

```python
model_rf = RandomForestClassifier(criterion='entropy')
model_rf.fit(x_train,y_train)
pred_rf = model_rf.predict(x_test)
```

```python
confusion_matrix(y_test,pred_rf)
```

```python
from sklearn.metrics import plot_confusion_matrix
ig, ax = plt.subplots( dpi = 100)
plot_confusion_matrix(model_rf,x_train,y_train, cmap ='viridis',ax=ax)
```

```python
accuracy= accuracy_score(y_test,pred_rf)
accuracy*100
```

### Decision Tree

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(x_train,y_train)
pred_dt = model_dt.predict(x_test)
```

```python
confusion_matrix(y_test,pred_dt)
```

```python
from sklearn.metrics import plot_confusion_matrix
ig, ax = plt.subplots( dpi = 100)
plot_confusion_matrix(model_dt,x_train,y_train, cmap ='viridis',ax=ax)
```

```python
accuracy= accuracy_score(y_test,pred_dt)
accuracy*100
```

### Logistic Regression

```python
model_lr = LogisticRegression()
model_lr.fit(x_train,y_train)
pred_lr = model_lr.predict(x_test)
```

```python
confusion_matrix(y_test,pred_lr)
```

```python
accuracy= accuracy_score(y_test,pred_lr)
accuracy*100
```

```python
from sklearn.metrics import plot_confusion_matrix
ig, ax = plt.subplots( dpi = 100)
plot_confusion_matrix(model_lr,x_train,y_train, cmap ='viridis',ax=ax)
```
